```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

OHE = OneHotEncoder(sparse_output=False)  # drop first to avoid redundancy
fuel_encoded = OHE.fit_transform(df[['Fuel_Type']])
df_encoded = pd.DataFrame(fuel_encoded, columns=OHE.get_feature_names_out(['Fuel_Type']))
```

```python
df_encoded.columns
```

```python
df.drop(columns=['Fuel_Type'],axis=0,inplace=True)
```

```python
df[['Fuel_Type_CNG', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol']] = df_encoded
```

```python
df
```

```python
df.drop(columns=['Car_Name'],axis=0,inplace=True)
```

```python
sns.heatmap(df.corr())
```

```python
x = df.drop(columns=['Selling_Price'],axis=0)
y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.2, random_state=43)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()

```

```python
RF.fit(x_train,y_train)
y_pred = RF.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,root_mean_squared_error,mean_squared_error

score = mean_squared_error(y_test,y_pred)
print("MSE: ", score)
cm1 = r2_score(y_test,y_pred)
print("R2 SCORE: ",cm1)
cm2 = mean_absolute_error(y_pred,y_test)
print("MAE: " ,cm2)
print("RMSE: ",root_mean_squared_error(y_test,y_pred))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6, edgecolors='k')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # y=x line

plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.legend()
plt.show()
```
